<h4><span style="color: #000000;">1. a)&nbsp;</span></h4>
<p>&nbsp;Generar una im&aacute;gen, mostrarla y explicar c&oacute;mo se gener&oacute;.</p>

```python
#imagen en negro 400x400
img = np.zeros((399,399,3),np.uint8)
#franja amarilla
#(R,G,B)=(255,255,0)
img[:,0:100,0] = 255
img[:,0:100,1] = 255
#franja cyan
#(R,G,B)=(0,255,255)
img[:,100:200,1] = 255
img[:,100:200,2] = 255
#franja roja
#(R,G,B)=(255,0,0)
img[:,200:300,0] = 255
#franja verde
#(R,G,B)=(0,255,0)
img[:,300:400,1] = 255

plt.imshow(img)
```

<h4><span style="color: #000000;">1. b)&nbsp;</span></h4>
<p>Lea la imagen "eiffel.jpg" que se encuentra en la carpeta "res", redimensionela a 800x600 p&iacute;xeles, muestre la imagen y su histograma asociado.</p>

```python
#Imagen original e histograma de los canales RGB
eiffel = cv2.imread('res/eiffel.jpg',cv2.IMREAD_COLOR)
eiffel = cv2.cvtColor(eiffel,cv2.COLOR_BGR2RGB)
eiffel = cv2.resize(eiffel,(800,600))
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(10,5))
colors = ('r','g','b')
fig.suptitle('Imagen e Histograma')
ax1.set_title('Imagen')
ax1.imshow(eiffel,aspect='auto')
ax2.set_title('Histograma')
for i, col in enumerate(colors):
    imgarray = eiffel[:,:,i].ravel()
    ax2.hist(imgarray,histtype='step',bins=255,
             range=(0.0,255.0),normed=True,color=colors[i])
```

<h4><span style="color: #000000;">1. a)&nbsp;</span></h4>
<p>Se implementa la transoformacin</p>

```python
#funcion que define la transformacion
def transform1(i1,i2,k):
    
    a = np.multiply(i1,1-k)
    b = np.multiply(i2,k)
    
    result = np.add(a,b)
    result = result.astype(np.uint8)
    return result
```

```python
#se leen las dos imagenes
#se transforman del espacio BGR a RGB
col = cv2.imread('res/col.png',cv2.IMREAD_COLOR)
col = cv2.cvtColor(col,cv2.COLOR_BGR2RGB)
plt.imshow(col)
```

```python
jap = cv2.imread('res/jap.png',cv2.IMREAD_COLOR)
jap = cv2.cvtColor(jap,cv2.COLOR_BGR2RGB)
plt.imshow(jap)
```

```python
#se debe reescalar la imagen porque tienen tamanos diferentes
col = cv2.resize(col,(250,166))
```

```python
#se realiza la transformacion

#parametro de la funcion
# k [0,1]
k = 0.5

transformada = transform1(col,jap,k)
fig, (ax1,ax2,ax3) = plt.subplots(1,3,figsize=(15,3))
ax1.set_title('Imagen 1')
ax2.set_title('Imagen 2')
ax3.set_title('Transformacion')
ax1.imshow(col)
ax2.imshow(jap)
ax3.imshow(transformada)
```

<p>Que ocurre cuando k vale 0, 0.5 y 1</p>

<h4><span style="color: #000000;">1. b)&nbsp;</span></h4>

```python
#se define la transformacion a*i + b
def transform2(i,a,b):
    s = np.multiply(i,a)
    result = np.add(s,b)
    result = result.astype(np.uint8)
    return result
```

```python
#parametros de la funcion
a = 10
b = -50
transformada = transform2(col,a,b)

fig, (ax1,ax3) = plt.subplots(1,2,figsize=(10,3))
ax1.set_title('Imagen')
ax3.set_title('Transformacion')
ax1.imshow(col)
ax3.imshow(transformada)
```
